```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import pandas as pd

%matplotlib inline
```

```python
with open('pdr_new.xgr') as fin, open('pdr.csv', 'w') as fout:
    o=csv.writer(fout)
    for line in fin:
        o.writerow(line.split())
```

```python
data = pd.read_csv('delay.csv')

data_f = data.as_matrix()
```

```python
y = data_f[:,0]
#x = np.arange(y.shape[0])

l = y.shape[0] // 10
y_avg = np.zeros((l))
for i in range(l):
    y_avg[i] = np.sum(y[i*10:(i+1)*10])


x = np.arange(y_avg.shape[0])
```

```python
fig = plt.figure()
plt.plot(x,y_avg)
fig.suptitle('Plot 1', fontsize=16)
plt.xlabel('Packets', fontsize=12)
plt.ylabel('Delay', fontsize=12)
plt.ylim(0,5.0)
fig.savefig('test1.jpg')
```

```python
data_throughput = pd.read_csv('throughput.csv')

data_t = data_throughput.as_matrix()
y = data_t[:,0]
x = data_t[:,1]
```

```python

fig = plt.figure()
plt.plot(x,y)
fig.suptitle('Plot 2', fontsize=16)
plt.xlabel('Time', fontsize=12)
plt.ylabel('Packet size', fontsize=12)
fig.savefig('throughput1.jpg')
```
